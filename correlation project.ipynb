```python
# First let's import the packages we will use in this project
# You can do this all now or as you need them
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None



# Now we need to read in the data
df = pd.read_csv(r'C:\Users\talha\Desktop\movies.csv')
```

```python
df.head()
```

```python
# We need to see if we have any missing data
# Let's loop through the data and see if there is anything missing

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
# It seems that a lot of data is missing from the budget column
# As we can not verify the data with missing columns we will drop the data with missing entries
df = df.dropna()
```

```python
#checking again for missing values
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
# Data Types for our columns

print(df.dtypes)
```

```python
#some entries have different release 'year' than the 'relaesed' columns
# to correct it we will make a new column named 'correctedyear' and extract the year from the released string
df['yearcorrect'] = df['released'].str.extract(pat = '([0-9]{4})').astype(int)
```

```python
df.head()
```

```python
#ordering by gross revenue
df=df.sort_values(by=['gross'], inplace= False , ascending= False)
```

```python
pd.set_option('display.max_rows', None)
```

```python
df.head()
```

```python
# Testing hypothesis that big budget equals big gross i.e. testing correlation
plt.scatter(x=df['budget'],y=df['gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel('Budget')
plt.ylabel('Gross Earnings')
```

```python
# plot the budget vs. gross using seaborn

sns.regplot(x='budget', y='gross',data=df, scatter_kws={"color":"red"}, line_kws={"color": "blue"})
```

```python
#lets check the correlation
df.corr(numeric_only=True)
```

```python
df.corr(numeric_only=True,method="pearson")
```

```python
#High correlation between budget and gross (0.7) , means the hypothesis was right
```

```python
#getting better visualization through heatmap
correlation_matrix=df.corr(numeric_only=True,method="pearson")
sns.heatmap(correlation_matrix,annot=True)
plt.title('Correlation Matrix for Numeric Features')

plt.show()
```

```python
# next lets see the correlation between company and gross
#first lets convert the non_numeric types to numeric
df_numerized= df
```

```python
df_numerized.head()
```

```python
for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name]= df_numerized[col_name].astype('category')
        df_numerized[col_name]= df_numerized[col_name].cat.codes
```

```python
df_numerized.head()
```

```python
#getting better visualization through heatmap
correlation_matrix=df_numerized.corr(numeric_only=True,method="pearson")
sns.heatmap(correlation_matrix,annot=True)
plt.title('Correlation Matrix for Numeric Features')

plt.show()
```

```python
df_numerized.corr(method="pearson")
```

```python
corr_mat = df_numerized.corr(method="pearson")
corr_pairs = corr_mat.unstack()
corr_pairs
```

```python
sorted_pairs= corr_pairs.sort_values()
sorted_pairs
```

```python
high_corr= sorted_pairs[(sorted_pairs) > 0.5]
high_corr
```
